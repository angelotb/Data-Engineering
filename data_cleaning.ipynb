```python
import pandas as pd

df = pd.read_csv('red_30_tech_us_sales.csv')
```

```python
df.columns = [col.strip().lower() for col in df.columns]
```

```python
df['order total'] = df['order total'].astype(str).str.replace(r'[\s"$#,]', '', regex=True).astype(float)
```

```python
print(df['order total'])
```

```python
df['price'] = df['price'].astype(str).str.replace(r'[\s"$#,]', '', regex=True).astype(float)
```

```python
df['discount'] = df['discount'].astype(str).str.replace(r'[\s"$#,]', '', regex=True).astype(float)
```

```python
df['orderdate'] = pd.to_datetime(df['orderdate'], format='%m/%d/%Y', errors='coerce')
```
